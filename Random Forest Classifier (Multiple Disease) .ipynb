# **‚≠ê Classification Model to Identify Multiple Disease**

```python
# import library
import pandas as pd
import numpy as np
```

```python
# import data
disease = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
# view data
disease
```

```python
# info of data
disease.info()
```

```python
# summary statistics
disease.describe()
```

```python
# check for missing value
disease.isnull().sum()
```

```python
# check for categories
disease.nunique()
```

```python
# correlation
disease.corr()
```

```python
# visualize pairplot
import seaborn as sns
sns.pairplot(disease,vars = ['itching','skin_rash','blackheads', 'scurring', 'skin_peeling', 'silver_like_dusting',
       'small_dents_in_nails', 'inflammatory_nails', 'blister',
       'red_sore_around_nose', 'yellow_crust_ooze'])
```

```python
# column names
disease.columns
```

```python
# define y
y = disease['prognosis']
```

```python
# define X
X = disease.drop(['prognosis'] , axis = 1)
```

```python
# split data
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state=2529,test_size=0.3)
```

```python
# verify shape
X_train.shape , X_test.shape  , y_train.shape  , y_test.shape 
```

```python
# select model
from sklearn.ensemble import RandomForestClassifier
model =  RandomForestClassifier()
```

```python
# train model
model.fit(X_train , y_train)
```

```python
# predict with model
y_pred = model.predict(X_test)
y_pred
```

```python
# model evaluation
```

```python
# model accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# model confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
# model classification report
from sklearn.metrics import classification_report
print(classification_report(y_test , y_pred))
```
